```python
from pyspark.sql.functions import *

df = spark.table("workspace.banking_db.silver_transactions")

# Customer monthly spending
gold_df = df.groupBy("Customer_ID", "year", "month") \
    .agg(
        sum("Transaction_Amount").alias("total_spent"),
        count("Transaction_ID").alias("total_transactions")
    )

gold_df.write.mode("overwrite").saveAsTable("workspace.banking_db.gold_customer_monthly")
```
