```python
# Read all 3 CSV files from Volume

bank_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/banking_db/banking_vol/bank_data.csv")

customer_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/banking_db/banking_vol/customer_data.csv")

transaction_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load("/Volumes/workspace/banking_db/banking_vol/transaction_data.csv")


# Save as Bronze tables
bank_df.write.mode("overwrite").saveAsTable("workspace.banking_db.bronze_bank")
customer_df.write.mode("overwrite").saveAsTable("workspace.banking_db.bronze_customer")
transaction_df.write.mode("overwrite").saveAsTable("workspace.banking_db.bronze_transaction")
```
